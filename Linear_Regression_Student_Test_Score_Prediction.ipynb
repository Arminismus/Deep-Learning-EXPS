```python
import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt

from tensorflow.keras.layers import Dense,Normalization

from tensorflow.keras import Sequential
import pandas as pd
```

```python
#import dataset:

#Student Test Score prediction:
data = pd.read_csv('/content/test_scores.csv')
```

```python
data
```

```python
data.drop('school',inplace = True,axis = 1)
```

```python
data.drop(['classroom','student_id'],inplace = True,axis = 1)
overhead = data
```

```python
def one_hot_encode(data,feature_values,feature):
  for value in feature_values:
    data[value] = (data[feature] == value)
    #make new values into ints:
    data[value] = data[value].replace({True:1,False:0})
  
  #remove original feature
  data.drop(feature,inplace = True,axis = 1)
```

```python
s = set()
for i in overhead['teaching_method']:
  if i not in s:
    s.add(i)
list(s)
```

```python
feature = 'teaching_method'
feature_values = ['Experimental','Standard']

def find_values(data,feature):
  s = set()
  for i in data[feature]:
    if i not in s:
      s.add(i)
  return list(s)

feature_values = find_values(feature= 'teaching_method',data = overhead)
```

```python
one_hot_encode(data = overhead,feature = 'school_setting',feature_values = find_values(overhead,'school_setting'))
```

```python
one_hot_encode(data = overhead,feature = 'teaching_method',feature_values = feature_values)
```

```python
one_hot_encode(data = overhead,feature = 'gender',feature_values = find_values(overhead,'gender'))
```

```python
overhead.drop('Male',inplace = True,axis = 1)
```

```python
overhead.drop('Non-public',inplace = True,axis = 1)
```

```python
y = overhead.pop('posttest').to_numpy()
```

```python
overhead = overhead.to_numpy()
```

```python
y
```

```python
#let's Create our Neural Network: No test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(overhead,y,test_size = 0.2,random_state = 12)
```

```python
x_train
```

```python
y_test
```

```python
#let's make the model and compile and fit it:
model = Sequential([
    Normalization(input_shape =(10,)),
    Dense(units = 14,activation = 'relu'),
    Dense(units = 3,activation = 'relu'),
    Dense(units = 1,activation = 'linear'),
])
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),loss = 'mse',metrics=['accuracy'])
```

```python
#make the Pandas dataframe into a vector
model.fit(x_train,y_train,epochs = 200)
```

```python
n = 0
model.predict(x_test[n]),y_test[n]
```

```python
model.evaluate(x_test,y_test)
```
