```python
import tensorflow as tf
from tensorflow.keras import Sequential

from tensorflow.keras.layers import Dense,Normalization
import numpy as np
import pandas as pd
```

```python
#let's Create the Artificial Data: Clean: One hot vectors  Noisy: Add noise to one hot vector

#ALL one hot 20 dimensional vectors:
vecs = [[1 if i==j else 0for i in range(20)] for j in range(20)]
```

```python
vecs = np.array(vecs)
```

```python
#Now let's create a data set where we sample one vector randomly:
dataset = []
for _ in range(1999):
  dataset.append(vecs[np.random.randint(0,19)])
```

```python
noisy_dataset = [[entry + (1e-3)*np.random.uniform() for entry in data] for data in dataset]
```

```python
x = np.array(noisy_dataset)
y = np.array(dataset)
```

```python
#let's make the model:
model = Sequential([
    tf.keras.Input(shape=(20,)),
    Dense(units = 20,activation = 'sigmoid'),
    Dense(units = 20,activation = 'sigmoid'),
])
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),loss = 'mse')
```

```python
y[0],x[0]
```

```python
model.fit(x,y,epochs = 200)
```

```python
model.predict(np.array([[1+(1e-3*np.random.uniform()) if i == 2 else 1e-3*np.random.uniform() for i in range(20)]]))
```
